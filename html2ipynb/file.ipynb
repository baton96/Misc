<h1 id="Running-Code">Running Code<a class="anchor-link" href="#Running-Code">¶</a></h1>

<p>First and foremost, the Jupyter Notebook is an interactive environment for writing and running code. The notebook is capable of running code in a wide range of languages. However, each notebook is associated with a single kernel.  This notebook is associated with the IPython kernel, therefore runs Python code.</p>

<h2 id="Code-cells-allow-you-to-enter-and-run-code">Code cells allow you to enter and run code<a class="anchor-link" href="#Code-cells-allow-you-to-enter-and-run-code">¶</a></h2>

<p>Run a code cell using <code>Shift-Enter</code> or pressing the <button class="btn btn-default btn-xs"><i class="icon-step-forward fa fa-play"></i></button> button in the toolbar above:</p>

```
a = 10
```

```
print(a)
```

<p>There are two other keyboard shortcuts for running code:</p>
<ul>
<li><code>Alt-Enter</code> runs the current cell and inserts a new one below.</li>
<li><code>Ctrl-Enter</code> run the current cell and enters command mode.</li>
</ul>

<h2 id="Managing-the-Kernel">Managing the Kernel<a class="anchor-link" href="#Managing-the-Kernel">¶</a></h2>

<p>Code is run in a separate process called the Kernel.  The Kernel can be interrupted or restarted.  Try running the following cell and then hit the <button class="btn btn-default btn-xs"><i class="icon-stop fa fa-stop"></i></button> button in the toolbar above.</p>

```
import time
time.sleep(10)
```

<h2 id="Restarting-the-kernels">Restarting the kernels<a class="anchor-link" href="#Restarting-the-kernels">¶</a></h2>

<p>The kernel maintains the state of a notebook's computations. You can reset this state by restarting the kernel. This is done by clicking on the <button class="btn btn-default btn-xs"><i class="fa fa-repeat icon-repeat"></i></button> in the toolbar above.</p>

<h2 id="sys.stdout-and-sys.stderr">sys.stdout and sys.stderr<a class="anchor-link" href="#sys.stdout-and-sys.stderr">¶</a></h2>

<p>The stdout and stderr streams are displayed as text in the output area.</p>

```
print("hi, stdout")
```

```
from __future__ import print_function
print('hi, stderr', file=sys.stderr)
```

<h2 id="Output-is-asynchronous">Output is asynchronous<a class="anchor-link" href="#Output-is-asynchronous">¶</a></h2>

<p>All output is displayed asynchronously as it is generated in the Kernel. If you execute the next cell, you will see the output one piece at a time, not all at the end.</p>

```
import time, sys
for i in range(8):
    print(i)
    time.sleep(0.5)
```

<h2 id="Large-outputs">Large outputs<a class="anchor-link" href="#Large-outputs">¶</a></h2>

<p>To better handle large outputs, the output area can be collapsed. Run the following cell and then single- or double- click on the active area to the left of the output:</p>

```
for i in range(50):
    print(i)
```

<p>Beyond a certain point, output will scroll automatically:</p>

```
for i in range(500):
    print(2**i - 1)
```
